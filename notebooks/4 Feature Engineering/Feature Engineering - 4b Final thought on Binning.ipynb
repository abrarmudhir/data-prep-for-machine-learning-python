### One Final Thought - Binning with the Target in Mind

Binning is a deliberate attempt to simplify our data, and can be helpful to reduce overfitting to the training data.

However, it is worth considering what predictive information we are sacrificing when we do this.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Data
data_df = pd.read_csv('binning_for_target.csv')
```

```python
#Plot the input feature as a histogram, and show the target as color.
sns.histplot(data = data_df, x = 'input', discrete = True, hue = 'target')
plt.show()
```

In this example, the decision boundary for the identification of B versus A exists clearly between 5 and 6.

Binning to maintain the shape of the distribution would combine columns 5 & 6, thus destroying some of the predictive information we have. The chart below shows the effect of binning in this way.

Values in the "5 & 6" bin could be A or B, but that information has been lost.

```python
#Plot the input feature as a histogram, and show the target as color.
sns.histplot(data = data_df, x = 'bins', discrete = True, hue = 'target', multiple='stack')
plt.show()
```
